```python
#Project f0r predecting salery of any person by taking "thr year of experience " as an independent variable.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
records = {'experience':['1','2','2.5','3','4','4.2','4.5','4.56','4.8','5'],
          'salery':['100000','130000','200000','220000','240000','300000','320000','323000','400000','430000']}
```

```python
records_frame = pd.DataFrame(records)
records_frame
```

```python
#divide datasets into x(independent) and y(dependent) variables
```

```python
x = records_frame.iloc[:,:-1].values
#before this: for row and after ,: for column.(-1) indicating we are lefting the last column for x.
```

```python
y = records_frame.iloc[:,1].values #(1) indicating we are taking only first column.
```

```python
#spliting datasets into training and test data set
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=3/5,random_state=0)
```

```python
#implement classifier for linear regression
#x_train
#y_train.reshape(4,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
simplelinearRegression = LinearRegression()
simplelinearRegression.fit(x_train,y_train)
```

```python
y_predict = simplelinearRegression.predict(x_train)
```

```python
y_predict #it will give the predicted value of salery according to experience.
```
